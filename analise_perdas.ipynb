```python
df_janeiro = pd.read_excel('/content/Relatório - Janeiro.xlsx')
df_fevereiro = pd.read_excel('/content/Relatório - Fevereiro.xlsx')
df_marco = pd.read_excel('/content/Relatório - Março.xlsx')
df_abril = pd.read_excel('/content/Relatório - Abril.xlsx')
df_maio = pd.read_excel('/content/Relatório - Maio.xlsx')
df_junho = pd.read_excel('/content/Relatório - Junho.xlsx')
df_julho = pd.read_excel('/content/Relatório - Julho.xlsx')
df_agosto = pd.read_excel('/content/Relatório - Agosto.xlsx')
df_setembro = pd.read_excel('/content/Relatório - Setembro.xlsx')
df_outubro = pd.read_excel('/content/Relatório - Outubro.xlsx')
```

```python
df_janeiro.fillna(0, inplace = True)
df_fevereiro.fillna(0, inplace = True)
df_marco.fillna(0, inplace = True)
df_abril.fillna(0, inplace = True)
df_maio.fillna(0, inplace = True)
df_junho.fillna(0, inplace = True)
df_julho.fillna(0, inplace = True)
df_agosto.fillna(0, inplace = True)
df_setembro.fillna(0, inplace = True)
df_outubro.fillna(0, inplace = True)
```

```python
# dataframes filtrados retirando-se as colunas NOME_PROPRIETARIO NOME_MORADOR ENDERECO COMPLEMENTO LOCALIZACAO

df_janeiro_data = df_janeiro.drop(['NOME_PROPRIETARIO', 'NOME_MORADOR', 'ENDERECO', 'COMPLEMENTO', 'LOCALIZACAO'], axis=1)
df_fevereiro_data = df_fevereiro.drop(['NOME_PROPRIETARIO', 'NOME_MORADOR', 'ENDERECO', 'COMPLEMENTO', 'LOCALIZACAO'], axis=1)
df_marco_data = df_marco.drop(['NOME_PROPRIETARIO', 'NOME_MORADOR', 'ENDERECO', 'COMPLEMENTO', 'LOCALIZACAO'], axis=1)
df_abril_data = df_abril.drop(['NOME_PROPRIETARIO', 'NOME_MORADOR', 'ENDERECO', 'COMPLEMENTO', 'LOCALIZACAO'], axis=1)
df_maio_data = df_maio.drop(['NOME_PROPRIETARIO', 'NOME_MORADOR', 'ENDERECO', 'COMPLEMENTO', 'LOCALIZACAO'], axis=1)
df_junho_data = df_junho.drop(['NOME_PROPRIETARIO', 'NOME_MORADOR', 'ENDERECO', 'COMPLEMENTO', 'LOCALIZACAO'], axis=1)
df_julho_data = df_julho.drop(['NOME_PROPRIETARIO', 'NOME_MORADOR', 'ENDERECO', 'COMPLEMENTO', 'LOCALIZACAO'], axis=1)
df_agosto_data = df_agosto.drop(['NOME_PROPRIETARIO', 'NOME_MORADOR', 'ENDERECO', 'COMPLEMENTO', 'LOCALIZACAO'], axis=1)
df_setembro_data = df_setembro.drop(['NOME_PROPRIETARIO', 'NOME_MORADOR', 'ENDERECO', 'COMPLEMENTO', 'LOCALIZACAO'], axis=1)
df_outubro_data = df_outubro.drop(['NOME_PROPRIETARIO', 'NOME_MORADOR', 'ENDERECO', 'COMPLEMENTO', 'LOCALIZACAO'], axis=1)

```
